```python
import cv2
import numpy as np

# Load the 1st image
image1 = cv2.imread("cars.jpg")

# Load the 2nd image
image2 = cv2.imread("humans.jpg")

def display_image(title, image):
  plt.imshow(image, cmap="gray")
  plt.title(title)
  plt.show()

# 2) Color Space Conversion
print("2) Color Space Conversion:")

# Grayscale
grayscale1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
grayscale2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
print("   - Grayscale:")
# Display the grayscale images
display_image("Grayscale Image1", grayscale1)
display_image("Grayscale Image2", grayscale2)

# HSV
hsv1 = cv2.cvtColor(image1, cv2.COLOR_BGR2HSV)
hsv2 = cv2.cvtColor(image2, cv2.COLOR_BGR2HSV)
print("   - HSV:")
# Display the HSV images
display_image("HSV Image 1", hsv1)
display_image("HSV Image 2", hsv2)
```

```python
# 3) Image Filtering
print("3) Image Filtering:")

# Gaussian Blur
blurred1 = cv2.GaussianBlur(image1, (5, 5), 0)  # Adjust kernel size as needed
blurred2 = cv2.GaussianBlur(image2, (5, 5), 0)  # Adjust kernel size as needed
print("   - Gaussian Blur:")
display_image("Gaussian Blurred Image 1", blurred1)
display_image("Gaussian Blurred Image 2", blurred2)

# Custom Averaging Kernel
kernel = np.ones((3, 3), np.float32) / 9  # Average weights
smoothed1 = cv2.filter2D(image1, -1, kernel)
smoothed2 = cv2.filter2D(image2, -1, kernel)
print("   - Custom Averaging Kernel:")
display_image("Custom Averaging Kernel Image 1", smoothed1)
display_image("Custom Averaging Kernel Image 2", smoothed2)
```

```python
print("4) Edge Detection:")

# Canny Edge Detection
edges1 = cv2.Canny(image1, 100, 200)
edges2= cv2.Canny(image2, 100, 200)

print("   - Canny Edge Detection:")
display_image("Canny Edge Detected Image1", edges1)
display_image("Canny Edge Detected Image2", edges2)

```

```python
# 6) Histogram
print("6) Histogram:")

def display_histogram(title, histogram):
    plt.figure()
    plt.title(title)
    plt.xlabel("Pixel Value")
    plt.ylabel("Frequency")
    plt.hist(histogram.ravel(), 256, [0, 256])
    plt.show()

# Compute the histogram
histogram1 = cv2.calcHist([image1], [0], None, [256], [0, 256])
histogram2 = cv2.calcHist([image2], [0], None, [256], [0, 256])

# Display the histogram
display_histogram("Histogram of Grayscale Image1", histogram1)
display_histogram("Histogram of Grayscale Image2", histogram2)
```
